```python
from kafka import KafkaConsumer
import json

# Créer le consumer Kafka
consumer = KafkaConsumer(
    "transactions",
    bootstrap_servers='localhost:9092',
    value_deserializer=lambda x: json.loads(x.decode('utf-8')),
    auto_offset_reset='earliest',
    group_id='transactions-group'
)

print("✅ Consumer connecté à Kafka !")

# Lire les transactions
count = 0
for message in consumer:
    transaction = message.value
    print(f"Reçu : {transaction['type']} - {transaction['amount']}$ - Fraude: {transaction['isFraud']}")
    count += 1
    if count >= 10:
        break

print("✅ 10 transactions reçues !")
```
