```python
import pandas as pd
from kafka import KafkaProducer
import json
import time

# Charger le dataset PaySim
df = pd.read_csv(r"C:\Users\HP\Desktop\kafka-spark-project\my_data\PS_20174392719_1491204439457_log.csv")

print(f"Dataset chargé : {len(df)} transactions")
print(df.head())
```

```python
# Créer le producer Kafka
producer = KafkaProducer(
    bootstrap_servers='localhost:9092',
    value_serializer=lambda x: json.dumps(x).encode('utf-8')
)

print("✅ Producer connecté à Kafka !")

# Envoyer les transactions une par une
for index, row in df.head(100).iterrows():
    transaction = {
        "step": int(row["step"]),
        "type": row["type"],
        "amount": float(row["amount"]),
        "nameOrig": row["nameOrig"],
        "nameDest": row["nameDest"],
        "isFraud": int(row["isFraud"])
    }
    producer.send("transactions", value=transaction)
    print(f"✅ Transaction envoyée : {transaction['type']} - {transaction['amount']}$")
    time.sleep(0.5)

producer.flush()
print("✅ Toutes les transactions envoyées !")
```
